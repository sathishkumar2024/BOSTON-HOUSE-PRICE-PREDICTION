```python
from sklearn import preprocessing
# Let's scale the columns before plotting them against MEDV
min_max_scaler = preprocessing.MinMaxScaler()
column_sels = ['lstat', 'indus', 'nox', 'ptratio', 'rm', 'tax', 'dis', 'age']
x = h_data.loc[:,column_sels]
y = h_data['medv']
x = pd.DataFrame(data=min_max_scaler.fit_transform(x), columns=column_sels)
fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(20, 10))
index = 0
axs = axs.flatten()
for i, k in enumerate(column_sels):
    sns.regplot(y=y, x=x[k], ax=axs[i])
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

#### This regression plot shows that maximum columns in data set having negative correlation with the target variable MDEV


#### And we remove the skewness of the data trough log transformation.

```python
## removing the skewness from the data
y =  np.log1p(y)
for col in x.columns:
    if np.abs(x[col].skew()) > 0.3:
        x[col] = np.log1p(x[col])
```

### splitting the data into input and target variables

```python
X = h_data.drop('medv',axis = 1) 
y = h_data.medv
```

```python
X
```

```python
y
```

## DATA PREPROCESSING

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X
```

```python
## split into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

## MODEL BUILDING

### Linear Regression

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
```

```python
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
## test the accuracy
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from math import sqrt

print('MAE:',mean_absolute_error(y_test, y_pred))
print('MSE :', mean_squared_error(y_test,y_pred))
print('RMSE :',sqrt(mean_squared_error(y_test,y_pred)))
print('R2_score :',r2_score(y_test, y_pred))
```

### Ridge Regression

```python
ridg = linear_model.Ridge(alpha=.5)
ridg.fit(X_train,y_train)
y_pred = ridg.predict(X_test)
```

```python
## model accuray
print('MAE:',mean_absolute_error(y_test, y_pred))
print('MSE :', mean_squared_error(y_test,y_pred))
print('RMSE :',sqrt(mean_squared_error(y_test,y_pred)))
print('R2_score :',r2_score(y_test, y_pred))
```

### XGBOOST

```python
import xgboost as xgb 
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,max_depth = 5, alpha = 10, n_estimators = 10) 
xg_reg.fit(X_train,y_train) 
y_pred = xg_reg.predict(X_test)
```

```python
## model accuray
print('MAE:',mean_absolute_error(y_test, y_pred))
print('MSE :', mean_squared_error(y_test,y_pred))
print('RMSE :',sqrt(mean_squared_error(y_test,y_pred)))
print('R2_score :',r2_score(y_test, y_pred))
```

### Randomforest 

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
## model accuray
print('MAE:',mean_absolute_error(y_test, y_pred))
print('MSE :', mean_squared_error(y_test,y_pred))
print('RMSE :',sqrt(mean_squared_error(y_test,y_pred)))
print('R2_score :',r2_score(y_test, y_pred))
```

#### Best parameter search for RandomForestRegressor using GridsearchCV because we got higher accuracy so we try to improve the model accuracy even higher


```python
# creating the parameters 
param = {'n_estimators': range(10, 101, 5),
              'criterion':['squared_error', 'absolute_error', 'poisson'],
             'max_features':['sqrt', 'log2', None]
}
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(regressor,param)
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
#appling the best parameters
random_reg = RandomForestRegressor(criterion ='poisson', max_features ='sqrt', n_estimators = 95)
random_reg.fit(X_train, y_train)
y_pred = random_reg.predict(X_test)
```

```python
print('MAE:',mean_absolute_error(y_test, y_pred))
print('MSE :', mean_squared_error(y_test,y_pred))
print('RMSE :',sqrt(mean_squared_error(y_test,y_pred)))
print('R2_score :',r2_score(y_test, y_pred))
```

#### its less score compare to what we get before using the gridsearchcv so we use the default parameters for this model
